```python
import h2o
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
h2o.init()
```

```python
x = [z*0.01 for z in range(1001)]
```

```python
print(x[0:5])
print(x[995:])
```

```python
# Prepare data
y = np.sin(x)
y = y + np.random.normal(0,0.1,len(x))
```

```python
# Visualize data
plt.plot(x,y)
```

```python
# Create pandas dataframe
sine_wave = pd.DataFrame({'a':x, 'b':y})
```

```python
# Convert dataframe to h2o frame
sine_wave_h2o = h2o.H2OFrame(sine_wave)
```

```python
sine_wave_h2o.summary()
```

```python
# To give name to frame
sine_wave_h2o = h2o.H2OFrame(
sine_wave,
destination_frame = "sine_wave"
)
```

```python
# To convert h2o frame to pandas dataframe use
sine_wave_h2o.as_data_frame()
```
